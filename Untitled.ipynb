```python
# -*- coding: utf-8 -*-
"""
Python Script for IoT Dataset Analysis and Neural Network Modeling
"""

# IMPORT REQUIRED LIBRARIES
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow import keras
import matplotlib.pyplot as plt

# LOAD THE CSV DATASET
file_path = 'data.csv'  # Replace with the correct path to your CSV file
data = pd.read_csv(file_path)

# CHECK DATA STRUCTURE
print("Dataset Overview:")
print(data.head())
print("\nDataset Shape:", data.shape)
print("Columns in Dataset:", data.columns)

# HANDLE MISSING DATA (Optional)
# Uncomment if needed to handle missing values
# data = data.dropna()  # Drop rows with missing values
# data = data.fillna(0)  # Fill missing values with 0

# EXTRACT FEATURES (X) AND TARGET (y)
X = data[['ts', 'co', 'humidity', 'light', 'gas', 'smoke', 'temp']].copy()
X['light'] = X['light'].astype(int)  # Convert boolean to integer
y = data['motion'].astype(int)  # Convert boolean target to integer

# VERIFY DATASET DIMENSIONS
assert len(X) == 405184, f"Expected 405184 rows, but got {len(X)}"
assert len(y) == len(X), f"Features and target lengths do not match: {len(X)} != {len(y)}"
assert X.shape[1] == 7, f"Expected 7 features, but got {X.shape[1]}"

# SPLIT DATA INTO TRAINING, VALIDATION, AND TEST SETS
validation_size = int(0.1 * len(X))  # 10% for validation
x_valid = X[:validation_size]
y_valid = y[:validation_size]

# Remaining data after validation split
X_remaining = X[validation_size:]
y_remaining = y[validation_size:]

# Split remaining data into training (70% of total) and test (20% of total)
test_size = int(0.2 * len(X))  # 20% of total for test
x_train = X_remaining[:-test_size]
y_train = y_remaining[:-test_size]
x_test = X_remaining[-test_size:]
y_test = y_remaining[-test_size:]

# VERIFY SPLIT SIZES
assert len(x_train) + len(x_test) + len(x_valid) == len(X), "Data split mismatch"
assert len(x_train) == len(y_train), "Training set size mismatch"
assert len(x_test) == len(y_test), "Test set size mismatch"
assert len(x_valid) == len(y_valid), "Validation set size mismatch"

print(f"Training set: {len(x_train)} samples")
print(f"Validation set: {len(x_valid)} samples")
print(f"Test set: {len(x_test)} samples")

# NORMALIZE THE DATA
x_valid = (x_valid - x_valid.min()) / (x_valid.max() - x_valid.min())
x_train = (x_train - x_train.min()) / (x_train.max() - x_train.min())
x_test = (x_test - x_test.min()) / (x_test.max() - x_test.min())

# DEFINE THE MODEL WITH DROPOUT AND REGULARIZATION
model = keras.models.Sequential([
    keras.Input(shape=[7]),  # Input layer
    keras.layers.Dense(300, activation="relu", kernel_regularizer=keras.regularizers.l2(0.01)),
    keras.layers.Dropout(0.3),  # Dropout layer
    keras.layers.Dense(100, activation="relu", kernel_regularizer=keras.regularizers.l2(0.01)),
    keras.layers.Dropout(0.3),  # Dropout layer
    keras.layers.Dense(1, activation="sigmoid")  # Output layer for binary classification
])

# COMPILE THE MODEL
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# EARLY STOPPING TO PREVENT OVERFITTING
early_stopping = keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=5, restore_best_weights=True
)

# TRAIN THE MODEL
history = model.fit(
    x_train, y_train,
    epochs=50,  # Increased epochs to use early stopping
    validation_data=(x_valid, y_valid),
    callbacks=[early_stopping]
)

# PLOT TRAINING AND VALIDATION ACCURACY
plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy', color='blue')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='green')
plt.title("Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# PLOT TRAINING AND VALIDATION LOSS
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss', color='orange')
plt.plot(history.history['val_loss'], label='Validation Loss', color='red')
plt.title("Training and Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# EVALUATE THE MODEL
print("Test Evaluation:")
test_results = model.evaluate(x_test, y_test)
print(f"Test Loss: {test_results[0]}")
print(f"Test Accuracy: {test_results[1]}")

# PREDICT ON NEW DATA
x_new = x_test[:3]  # Take first 3 samples from test data
y_prob = model.predict(x_new)
y_pred = (y_prob > 0.5).astype(int)  # Convert probabilities to binary predictions
print("Predictions for first 3 samples:", y_pred.flatten())
print("Actual values:", y_test[:3].values)
```
